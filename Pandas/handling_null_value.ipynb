```python
import pandas as pd
import numpy as np
df=pd.read_csv("expense_data_1.csv")
df
```

```python
df.isnull().sum()
```

```python
#delete null value 
print(df.dropna(how='all'))
```

```python
print(df.dropna(thresh=1))
print(df.dropna(subset=['Subcategory','Note.1']))


```

```python
#replace null value /NaN value by any  value 
print(df.replace(np.nan,"hii"))
```

```python
df['Subcategory']=df['Subcategory'].replace(np.nan,0)
```

```python
# for categorical data use upper fill and lower fill
#use fillna () method
data1={'id':[10,20,300],'name':['ansar ','nan','ashu'],'age':[21,22,20]}
print(data1)

```

```python
data2=pd.DataFrame(data1)
data3=data2.ffill(axis=1)
print(data3)
```

```python
 # col transformation using loc 
df.loc[df['Account.1']>200,"Account.2"]="midum"
df.loc[df['Account.1']>100 ,'Account.2']="high"
df.loc[df['Account.1']<=100,"Account.2"]="low"
print(df)
```

```python
df["INR Percentage"]=(df['INR']/100)*20
print(df)
```

```python
#map Function 
months={"months":["janaury","febray","march","april","may","june"]}
mon=pd.DataFrame(months)
print(mon)
```

```python
def addcol_to_substring(value):
    return value[:3]
mon['short_month']=mon['months'].map(addcol_to_substring)
print(mon)
```
